```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_transfermarkt_players(urls, headers=None, cookies=None):

    all_records = []

    for url in urls:
        # 1) On déduit l'ID du joueur à partir de l'URL
        player_id = url.split('/')[-1]

        # 2) Requête sur la page HTML du profil
        response = requests.get(url, headers=headers, cookies=cookies)
        if response.status_code != 200:
            print(f"Erreur {response.status_code} sur l'URL : {url}")
            continue

        soup = BeautifulSoup(response.content, "html.parser")

        # Nom du joueur
        try:
            name_raw = soup.select_one('h1.data-header__headline-wrapper').text
            player_name = name_raw.split('\n')[-1].strip()
        except AttributeError:
            player_name = None

        # Numéro du joueur
        try:
            number_raw = soup.select_one('span.data-header__shirt-number').text
            player_number = number_raw.strip().replace('#', '')
        except AttributeError:
            player_number = None

        # Valeur marchande
        try:
            value_elem = soup.select_one('a.data-header__market-value-wrapper')
            raw_value = value_elem.get_text(separator=' ', strip=True)
            # Exemple : "30,00  mio. € Dernier changement: 10 déc. 2024"
            # On isole la partie avant "Dernier changement:"
            if "Dernier changement:" in raw_value:
                raw_value = raw_value.split("Dernier changement:")[0].strip()
            player_value = raw_value
        except AttributeError:
            player_value = None

        # Club
        try:
            club_elem = soup.select_one('span.data-header__club a')
            player_club = club_elem.text.strip()
        except AttributeError:
            player_club = None

        # Ligue
        try:
            league_elem = soup.select_one('span.data-header__league a')
            player_league = league_elem.text.strip()
        except AttributeError:
            player_league = None

        # Position
        position_tag = soup.find('dd', class_='detail-position__position')
        position = position_tag.text.strip() if position_tag else 'Unknown'

        #Durée Contrat
        contrat_jusqua_tag = soup.find('span', string=lambda s: s and "Contrat jusqu" in s)
        date_tag = contrat_jusqua_tag.find_next('span',class_='info-table__content info-table__content--bold') if contrat_jusqua_tag else None
        date_fin_contrat = date_tag.text.strip() if date_tag else "Unknown"



        # Age
        age_tag = soup.find('span', string='Naissance (âge):')
        age_value = age_tag.find_next('span', class_='info-table__content info-table__content--bold') if age_tag else None
        age = age_value.text.strip() if age_value else 'Unknown'

        # --- API performance ---
        performance_url = f"https://www.transfermarkt.us/ceapi/player/{player_id}/performance"
        try:
            performance_data = requests.get(performance_url, headers=headers, cookies=cookies).json()
        except Exception as e:
            print(f"performance erreur pour {player_name} : {e}")
            performance_data = []

        # performance_data est en général une liste d'objets
        for i in performance_data:
            record = {
                "player_id": player_id,
                "player_name": player_name,
                "player_number": player_number,
                "player_value": player_value,
                "player_club": player_club,
                "player_league": player_league,
                "position": position,
                "age": age,
                "contract_end": date_fin_contrat,

                # Données performance
                "competition": i.get("competitionDescription"),
                "season": i.get("nameSeason"),
                "gamesPlayed": i.get("gamesPlayed"),
                "goalsScored": i.get("goalsScored"),
                "assists": i.get("assists"),
                "yellowCards": i.get("yellowCards"),
                "redCards": i.get("redCards"),
                "minutesPlayed": i.get("minutesPlayed"),
            }
            all_records.append(record)

    # Conversion en DataFrame
    df = pd.DataFrame(all_records)
    return df
```

Scraping sur la page transfert

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

def scrape_transfermarkt_transfers_selenium(urls):
    """
    Récupère le 'transfer_fee' (premier montant trouvé)
    pour chaque joueur, en utilisant Selenium (rendu JS).
    """
    # On configure les options Selenium
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("start-maximized")
    options.add_argument("enable-automation")
    options.add_argument("--disable-infobars")
    options.add_argument("--disable-dev-shm-usage")
    driver = webdriver.Chrome(options=options)

    results = []

    for url in urls:
        # 1) Corriger l’URL si elle contient "/profil/"
        if "/profil/" in url:
            url = url.replace("/profil/", "/transfers/")

        # 2) On extrait l'ID du joueur pour la suite
        player_id = url.split('/')[-1]

        # 3) Charger la page
        driver.get(url)
        # Attendre que la page se charge
        time.sleep(1)

        # 4) Récupérer le HTML après rendu JS
        html_content = driver.page_source

        # 5) Parse avec BeautifulSoup
        soup = BeautifulSoup(html_content, "html.parser")

        # 6) Recherche de la première ligne de transfert
        try:
            transfer_rows = soup.find_all('div', class_='grid tm-player-transfer-history-grid')
            if transfer_rows:
                first_row = transfer_rows[0]
                first_transfer_fee_tag = first_row.find(
                    'div',
                    class_='tm-player-transfer-history-grid__fee'
                )
                first_transfer_fee = (first_transfer_fee_tag.text.strip()
                                      if first_transfer_fee_tag
                                      else 'Unknown')
            else:
                first_transfer_fee = 'No transfer rows found'
        except Exception as e:
            first_transfer_fee = f"Error: {e}"

        # 7) On stocke le résultat dans un dictionnaire
        results.append({
            "player_id": player_id,
            "transfer_fee": first_transfer_fee
        })

    # 8) Fermer le navigateur UNE FOIS
    driver.quit()

    # 9) Convertir en DataFrame pour un usage plus pratique
    df = pd.DataFrame(results)
    return df

```

```python
df=pd.read_csv("df_final1.csv",delimiter=",")
```

```python
df=df.drop(columns=["player_id","season"])

df["minutesPlayed_CoupeFrance"] = df.apply(lambda row: row["minutesPlayed"] if row["competition"] == "Coupe de France" else 0, axis=1)
df["minutesPlayed_UEFA"] = df.apply(lambda row: row["minutesPlayed"] if row["competition"] == "UEFA Champions League" else 0, axis=1)
df["goalsScored_UEFA"] = df.apply(lambda row: row["goalsScored"] if row["competition"] == "UEFA Champions League" else 0, axis=1)
```

```python
df = df.groupby("player_name").agg({
    "player_number": "first",
    "player_value": "first",
    "player_club": "first",
    "gamesPlayed": "sum",
    "goalsScored": "sum",
    "assists": "sum",
    "yellowCards": "sum",
    "redCards": "sum",
    "minutesPlayed": "sum",
    "minutesPlayed_UEFA": "sum",
    "goalsScored_UEFA": "sum",
    "position": "first",
    "age": "first",
    "player_league": "first",
    "transfer_fee": "first",
    "contract_end": "first",
}).reset_index()
```

```python
def convert_player_value(value):
    value = value.replace(" €", "").replace(",", ".").replace("mio.", "e6").replace("K", "e3").replace(" ", "")
    return int(float(eval(value)))

df["player_value"] = df["player_value"].apply(convert_player_value)
```

```python
def clean_transfer_fee(value):
    if isinstance(value, str):
        value = value.replace(",", ".")  # Remplacer les virgules par des points
        if "mio" in value:
            return float(value.split()[0]) * 1_000_000  # Convertir "mio" en millions
        elif "K" in value:
            return float(value.split()[0]) * 1_000  # Convertir "K" en milliers
        elif value.strip() == "-" or value.strip().lower() == "nan":
            return 0  # Gérer les "-" et "NaN" comme 0
    return 0

# Appliquer la fonction à la colonne
df["transfer_fee"] = df["transfer_fee"].apply(clean_transfer_fee)
```

```python
df['age'] = df['age'].str.extract(r'\((\d+)\)')
df
```

```python
df["remaining_years"] = df["contract_end"].apply(lambda date: int(date.split()[-1]) - 2025)
df=df.drop(columns=["contract_end"])
```

```python
df['player_value'] = df['player_value'].astype(int)
df['gamesPlayed'] = df['gamesPlayed'].astype(int)
df['goalsScored'] = df['goalsScored'].astype(int)
df['assists'] = df['assists'].astype(int)
df['yellowCards'] = df['yellowCards'].astype(int)
df['redCards'] = df['redCards'].astype(int)
df['minutesPlayed'] = df['minutesPlayed'].astype(int)
df['minutesPlayed_UEFA'] = df['minutesPlayed_UEFA'].astype(int)
df['goalsScored_UEFA'] = df['goalsScored_UEFA'].astype(int)
df['age'] = df['age'].astype(int)
df['transfer_fee'] = df['transfer_fee'].astype(int)
df['remaining_years'] = df['remaining_years'].astype(int)
```

# Prediction

```python
kfold = KFold(n_splits=7, shuffle=True, random_state=42)

# Modèle initialisé
#model = LinearRegression()
model = Ridge(alpha=2.0)

# Listes pour stocker les résultats des métriques R²
train_scores = []
test_scores = []

for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    train_scores.append(train_r2)
    test_scores.append(test_r2)
```

## Evaluation score

```python
mean_train_r2 = np.mean(train_scores)
std_train_r2 = np.std(train_scores)
mean_test_r2 = np.mean(test_scores)
std_test_r2 = np.std(test_scores)

print("The model performance for training set")
print("--------------------------------------")
print('Mean R2 score is {:.4f} ± {:.4f}'.format(mean_train_r2, std_train_r2))

print("\nThe model performance for testing set")
print("--------------------------------------")
print('Mean R2 score is {:.4f} ± {:.4f}'.format(mean_test_r2, std_test_r2))
```

```python
resid_train = y_train - y_train_pred

# Pour l'ensemble de test
resid_test = y_test - y_test_pred

# Tracé des résidus pour l'ensemble d'entraînement
plt.scatter(y_train_pred, resid_train, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valeurs prédites')
plt.ylabel('Résidus')
plt.title('train')
plt.show()

plt.scatter(y_test_pred, resid_test, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Valeurs prédites')
plt.ylabel('Résidus')
plt.title('test')
plt.show()
```
